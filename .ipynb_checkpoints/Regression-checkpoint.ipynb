```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
sns.set_context('paper', font_scale=1.4)
```

```python
df = pd.read_csv('data/superstore-sales.csv', encoding='unicode_escape')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Month Year'] = df['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
df['TS Date'] = df['Order Date'].apply(lambda x: x.strftime('%Y%m'))
df.head()
```

```python
df.info()
```

```python
ts = df[['TS Date', 'Sales']].groupby(
    by='TS Date', 
    as_index=False
).sum().sort_values(
    by='TS Date',
    ascending=True
)
```

```python
plt.figure(figsize=(18, 8))
plt.xticks(rotation=45)

sns.lineplot(
    x='TS Date', 
    y='Sales', 
    data=ts
);
```
